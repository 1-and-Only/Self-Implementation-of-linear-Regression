```python
import numpy as np
import pandas as pd
```

```python
data=np.genfromtxt("train.csv",delimiter=",")
pd.DataFrame(data)
```

```python
x_train=data[:,0:len(data[0])-1]#features
x_train_x=x_train
c=[[1]]*len(x_train)#extra feature to take care of c
x_train=np.append(x_train,c,axis=1)
y_train=data[:,len(data[0])-1:len(data[0])]#output
#score is coeeficent of determination defined as 1-u/v
# we need to give value of y_pred in .csv format
pd.DataFrame(x_train),pd.DataFrame(y_train)
```

```python
def step_gradient(m,points,y_train,learning_rate):
    m_slope=np.zeros(len(points[0]))
    N=len(points)
    for i in range(N):
        x=points[i,:]
        y=y_train[i,:]
        for j in range(len(x)):
            m_slope[j]=m_slope[j]+((-2/N)*((y-(m*x).sum())*x[j]))
    new_m = m- learning_rate*m_slope
    return new_m
```

```python
def gd(m,x_train,y_train,learning_rate,num_iterations):
    for i in range(num_iterations):
        m=step_gradient(m,x_train,y_train,learning_rate)
        print(i,"cost:",cost(x_train,y_train,m))
    return m
```

```python
def cost(points,y_train,m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        total_cost += (y_train[i]-(points[i,:]*m).sum())**2
    return total_cost/M
```

```python
def score(y_true,y_pred):
    u=(y_true-y_pred).sum()
    print("u:",u,end="")
    v=(y_true-y_true.mean()).sum()
    print("v:",v)
    return 1-(u/v)
```

```python
# now since the features are more than 1 we need to use gradient decentend,so fit will give me a array of values of m
def fit(x_train,y_train):#we need to provide x_train and y_train as input
    m=np.zeros(len(x_train[0]))
    learning_rate=0.000000935
    num_iterations=100
    final_slope=gd(m,x_train,y_train,learning_rate,num_iterations)
    print(final_slope)
    return final_slope
    
def predict(m):
    x_test=np.genfromtxt("test.csv",delimiter=',')
    print(pd.DataFrame(x_test))
    y_pred=np.zeros(len(x_test))
    c=m[len(m)-1:len(m)]
    m=m[0:len(m)-1]
    for i in range(len(x_test)):
        x=x_test[i,:]
        y_pred[i]=(m*x).sum()+c
    return y_pred

m=fit(x_train,y_train)
y_pred=predict(m)
```
